## Как отфильтровать определенные строки из `DataFrame`?
Аналогично `WHERE` in SQL

```python
# Например нам нужны пассажиры старше 35 лет
above_35 = titanic[titanic['Age']>35]
above_35
```

Будут выбраты те строки которые `True` в предикате `titanic['Age']>35`

```python
(titanic['Age']>35).value_counts()
```

Аналогично запросу в `SQL`: `SELECT * FROM titanic
                             WHERE Pclass IN (2,3)`

```python
class_23 = titanic[titanic['Pclass'].isin([2,3])]
class_23
```

```python
#equivalent
class_23 = titanic[(titanic["Pclass"] == 2) | (titanic["Pclass"] == 3)]
class_23
```

При объединении нескольких условных операторов каждое условие должно быть заключено в круглые скобки (). Кроме того, вы не можете использовать `or/and`, но вам нужно использовать оператор или `|` и оператор `&`.

```python
age_exists = titanic[titanic['Age'].notna()]
age_exists
```

`notna()` проверка на отсутсвие `null` в строчке

```python
name_above_35 = titanic.loc[titanic['Age']>35, ['Name', 'Age']]
name_above_35
```

В `loc`/`iloc` первым параметром передаются `строки` которые мы хотим выбрать, после запятой идут `поля` которые мы хотим выбрать


При использовании имен столбцов, меток строк или выражения условия используйте оператор loc перед скобками выбора []. Для части до и после запятой вы можете использовать одну метку, список меток, часть меток, условное выражение или двоеточие. Использование двоеточия указывает, что вы хотите выбрать все строки или столбцы.

Тесно связан с `pivot` метод `stack` и `unstack`. Эти методы преднахзначены для тесной работы с `MultiIndex`
Что делают `stack` and `unstack`:
- `stack` "поворачивает" колонку в намы низкий уровень индекса
- `unstack` соответствует всоему заванияделает обратное `stack`

```python
tuples = list(zip(*[
    ["bar", "bar", "baz", "baz", "foo", "foo", "qux", "qux"],
    ['one', 'two', 'one', 'two', 'one', 'two', 'one', 'two']
                    ]
                 )
             ) 
index = pd.MultiIndex.from_tuples(tuples, names=['first', 'second'])
df = pd.DataFrame(np.random.randn(8,2), index=index)
df
```

```python
df2 = df[:4]
stacked = df2.stack()
stacked # появился новый индекс последнео уровня
```

```python
stacked.unstack() # возвращает в исходное положение
```

```python
stacked.unstack(1) # возвращает следующий уровень
```

    https://pandas.pydata.org/pandas-docs/stable/_images/reshaping_unstack_0.png - unstack(0)

```python
stacked.unstack("second")
```

`stack()` И `unstack()` методы неявно сортируют соответствующие уровни индекса. Отсюда и призыв к `stack()`а потом `unstack()`или наоборот, это приведет к отсортированной копии оригинала `DataFrame` или `Series`:

# Отличия `string` от `object`
- в `string` если использовался какой то строкойвый метод который всегда возвращает число, например, `count()` то тип строки всегда превращается в `int` вне зависимости от наличия `Nan`
- в `object` же возвращается `int` когда там отсутствует `Nan` в противном случае он вернет `float`

```python
s = pd.Series(['a','b','c', np.nan, 'd'], dtype='string')
s.str.count('a')
```

```python
s.dropna().str.count("a")
```

```python
s_ob = pd.Series(['a',None,'b'])
s_ob.str.count('a') # float64
```

```python
s_ob.dropna().str.count('a') #  int
```

    Некоторые строковые методы, такие как Series.str.decode(), недоступны в StringArray, поскольку StringArray содержит только строки, а не байты.
    В операциях сравнения arrays.StringArray и Series, поддерживаемые StringArray, будут возвращать объект с BooleanDtype, а не объект bool dtype. Отсутствующие значения в StringArray будут распространяться в операциях сравнения, а не всегда сравнивать неравные, как numpy.nan.
    Все остальное, что следует в оставшейся части этого документа, в равной степени относится к строковому и объектному dtype.
